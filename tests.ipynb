```python
import minari 

dataset = minari.load_dataset("D4RL/pointmaze/large-v2")
```

```python
sample = list(dataset.sample_episodes(n_episodes=5)) # this is how to sample episodes 
print("Sample IDs:", [ep.id for ep in sample])
```

```python
env = dataset.recover_environment(render_mode="rgb_array")
env.reset()
img = env.render()  # No mode argument here
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
env.close()
```
